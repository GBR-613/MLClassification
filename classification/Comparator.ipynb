```python
import glob
import os
import pickle
from collections import namedtuple 
from pathlib import Path
from operator import itemgetter
from IPython.core.display import display, HTML

DocInfo = namedtuple('DocInfo', 'name actLabs predLabs dtype')
CategoryInfo = namedtuple('CategoryInfo', 'name actLabs predLabs wrongLabs notPredLabs qtyDocs qtyPredDocs precision recall f1')
ModelInfo = namedtuple('ModelInfo', 'name dataSet cats qtyDocs qtyPredDocs qtyPartDocs qtyWrongDocs actLabs predLabs wrongLabs notPredLabs emr accuracy precision recall f1 hl macroPrecision macroRecall macroF1 microPrecision microRecall microF1 categories docs')

homePath = str(Path.home()) + "/MLClassificationData"
modelInfoPath = homePath + '/models/rtanews/modelinfo/'
pkls = []
os.chdir(modelInfoPath)
for fp in glob.glob("*"):
    print("Open file %s"%(fp))
    with open(fp, 'rb') as f:
        data = pickle.load(f)        
        #sorted(data.docs, key=itemgetter(0))
        pkls.append(data)
print("Load %d model info file(s)"%(len(pkls)))
```
